```python
# Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import numpy as np
import matplotlib.pyplot as plt

low, high = 1, 6  # For example, a six-sided die

outcomes = np.arange(low, high + 1)

probabilities = np.ones(len(outcomes)) / len(outcomes)

plt.figure(figsize=(8, 6))
plt.bar(outcomes, probabilities, width=0.5, color='skyblue', edgecolor='black')

plt.title('Probability Mass Function (PMF) of Discrete Uniform Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks(outcomes)
plt.ylim(0, 1)
plt.show()
```

```python
# Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):

    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1 for a Bernoulli distribution.")

    if x == 1:
        return p
    elif x == 0:
        return 1 - p

p = 0.7
x = 1
pdf_value = bernoulli_pdf(x, p)
print(f"The probability of x = {x} for a Bernoulli distribution with p = {p} is {pdf_value:.2f}")
```

```python
# Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

import numpy as np
import matplotlib.pyplot as plt

low, high = 1, 6

outcomes = np.arange(low, high + 1)

probabilities = np.ones(len(outcomes)) / len(outcomes)

cdf = np.cumsum(probabilities)

plt.figure(figsize=(8, 6))
plt.step(outcomes, cdf, where='post', label="CDF", color='skyblue', linewidth=2)

plt.title("Cumulative Distribution Function (CDF) of Discrete Uniform Distribution")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.xticks(outcomes)
plt.yticks(np.linspace(0, 1, 11))
plt.grid(True)

plt.show()
```

```python
# Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    data = np.array(data)


    mean = np.mean(data)
    std_dev = np.std(data)


    z_scores = (data - mean) / std_dev
    return z_scores

def plot_z_scores(data):
    z_scores = calculate_z_scores(data)

    plt.figure(figsize=(8, 6))
    plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7, density=True)


    plt.title("Z-Scores of the Dataset")
    plt.xlabel("Z-Score")
    plt.ylabel("Frequency")
    plt.grid(True)


    plt.show()

data = np.random.normal(0, 1, 1000)

plot_z_scores(data)
```

```python
#Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

import numpy as np
import matplotlib.pyplot as plt


sample_size = 30
num_samples = 1000
mean = 50
std_dev = 10


def central_limit_theorem_normal(sample_size, num_samples, mean, std_dev):

    sample_means = []


    for _ in range(num_samples):
        sample = np.random.normal(mean, std_dev, sample_size)
        sample_means.append(np.mean(sample))

    return sample_means


sample_means = central_limit_theorem_normal(sample_size, num_samples, mean, std_dev)

sample_means
```

```python
#  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):

    mean = np.mean(data)
    std_dev = np.std(data)

    z_score = (x - mean) / std_dev
    return z_score

def plot_standard_normal_distribution_and_z(data, x):

    z_score = calculate_z_score(data, x)

    x_values = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x_values, 0, 1)

    plt.figure(figsize=(8, 6))
    plt.plot(x_values, pdf, label='Standard Normal Distribution (μ=0, σ=1)', color='blue', linewidth=2)

    plt.axvline(z_score, color='red', linestyle='dashed', label=f'Z-score: {z_score:.2f}')

    plt.title(f"Z-score for Data Point {x}")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)


    plt.show()

    return z_score


data = np.random.normal(50, 10, 1000)
x = 60

z_score = plot_standard_normal_distribution_and_z(data, x)

print(f"The Z-score for the data point {x} is: {z_score:.2f}")
```

```python
# mplement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm


def z_test(sample_mean, population_mean, std_dev, sample_size, alpha=0.05):



    z_statistic = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))

    critical_value = norm.ppf(1 - alpha / 2)

    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"Critical value: ±{critical_value:.4f}")

    if abs(z_statistic) > critical_value:
        print("Reject the null hypothesis: There is a significant difference.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference.")

sample_mean = 5.6
population_mean = 5.5
std_dev = 0.8
sample_size = 30


z_test(sample_mean, population_mean, std_dev, sample_size)
```

```python
# Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)

    n = len(data)


    z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    standard_error = sample_std / np.sqrt(n)


    margin_of_error = z_value * standard_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound


data = np.random.normal(50, 10, 100)


lower, upper = confidence_interval(data, confidence_level=0.95)


print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
# Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
import scipy.stats as stats


def confidence_interval(data, confidence_level=0.95):


    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)


    n = len(data)


    z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)


    standard_error = sample_std / np.sqrt(n)


    margin_of_error = z_value * standard_error


    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound


np.random.seed(0)
true_mean = 50
true_std = 10
sample_size = 100


data = np.random.normal(true_mean, true_std, sample_size)


lower, upper = confidence_interval(data, confidence_level=0.95)


print(f"95% Confidence Interval for the mean: ({lower:.2f}, {upper:.2f})")


print("\nInterpretation:")
print(f"We are 95% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")
```

```python
# Write a Python program to compare the standard deviations of two datasets and visualize the difference.
import numpy as np
import matplotlib.pyplot as plt


def compare_standard_deviations(data1, data2):

    std_dev1 = np.std(data1, ddof=1)
    std_dev2 = np.std(data2, ddof=1)


    print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")


    plt.figure(figsize=(10, 6))


    plt.subplot(1, 2, 1)
    plt.hist(data1, bins=20, alpha=0.7, color='blue', edgecolor='black')
    plt.title(f'Dataset 1\nStd Dev = {std_dev1:.2f}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)


    plt.subplot(1, 2, 2)
    plt.hist(data2, bins=20, alpha=0.7, color='green', edgecolor='black')
    plt.title(f'Dataset 2\nStd Dev = {std_dev2:.2f}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)


    plt.tight_layout()
    plt.show()


np.random.seed(0)
data1 = np.random.normal(loc=50, scale=5, size=500)
data2 = np.random.normal(loc=50, scale=15, size=500)


compare_standard_deviations(data1, data2)
```

```python
# Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
import numpy as np


def calculate_range_iqr(data):

    data_range = np.max(data) - np.min(data)


    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    return data_range, IQR


np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=1000)

data_range, IQR = calculate_range_iqr(data)


print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {IQR:.2f}")
```

```python
#  Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt


def z_score_normalization(data):

    mean = np.mean(data)
    std_dev = np.std(data)


    normalized_data = (data - mean) / std_dev

    return normalized_data, mean, std_dev


np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=1000)


normalized_data, mean, std_dev = z_score_normalization(data)


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title('Original Data (Before Normalization)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)


plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, alpha=0.7, color='green', edgecolor='black')
plt.title('Normalized Data (Z-score Normalization)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)


plt.tight_layout()
plt.show()


print(f"Original Mean: {mean:.2f}, Original Std Dev: {std_dev:.2f}")
print(f"Normalized Mean: {np.mean(normalized_data):.2f}, Normalized Std Dev: {np.std(normalized_data):.2f}")
```

```python
#  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
import scipy.stats as stats


def calculate_skewness_kurtosis(data):

    skewness = stats.skew(data)


    kurtosis = stats.kurtosis(data)

    return skewness, kurtosis


np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=1000)


skewness, kurtosis = calculate_skewness_kurtosis(data)


print(f"Skewness of the dataset: {skewness:.2f}")
print(f"Kurtosis (Excess) of the dataset: {kurtosis:.2f}")
```
